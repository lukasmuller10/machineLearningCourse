```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
data = pd.read_excel('D02_Boston.xlsx')
data.head()
```

```python
data= data.iloc[:,1:]
data.head()
```

```python
#dispersion diagram
columns = data.columns
for col in columns:
    print('%s pearsonr'%col ,'= %.3f'% pearsonr(data[col], data['target'])[0] ) #'correlation'
    data.plot.scatter(x = col, y = "target")
```
